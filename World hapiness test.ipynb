```python
data_2017.rename(columns={
    'Happiness.Rank': 'Hapiness_Rank',
    'Happiness.Score': 'Hapiness_Score',
    'Whisker.high' : 'Whisker_High',
    'Whisker.low': 'Whisker_Low',
    'Economy..GDP.per.Capita.': 'Economy_GDP_Per_Capita',
    'Health..Life.Expectancy.':'Health_Life_Expectency',
    'Trust..Government.Corruption.':'Trust_Government_Corruption',
    'Dystopia.Residual':'Dystopia_Residual'    
},inplace = True)
```

```python
data_2017.head(10)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data_2017.Health_Life_Expectency,data_2017.Hapiness_Score,color = 'purple')
plt.xlabel(' Life Expectancy')
plt.ylabel('Happiness')
plt.title('hapiness vs life expectency')

plt.figure(figsize=(10,5))
plt.plot(data_2017.Health_Life_Expectency,data_2017.Hapiness_Score,color = 'purple')
plt.xlabel('Life Expectancy')
plt.ylabel('Happiness')
plt.title('hapiness vs life expectency')


```

```python
plt.figure(figsize=(10,5))
plt.scatter(data_2017.Trust_Government_Corruption,data_2017.Freedom, color = 'red')
plt.xlabel('Trust_Government_Corruption')
plt.ylabel('freedom')
plt.title('trust in government (corruption) vs freedom')




```

```python
plt.figure(figsize=(10,10))
plt.plot(data_2017.Trust_Government_Corruption, data_2017.Freedom, color = 'red', marker ='o',linestyle='-.')
plt.xlabel('Trust_Government_Corruption')
plt.ylabel('freedom')
plt.title('trust in government (corruption) vs freedom')
```

```python
plt.figure(figsize=(10,5))
data_2017.Freedom.plot(kind = 'line',color = 'red', label = 'freedom', linestyle = ':')
data_2017.Generosity.plot(kind = 'line',color = 'black', label = 'generosity', linestyle = ':')
plt.legend(loc='upper right')
plt.title('freedom vs generosity')
plt.xlabel()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data_2017.Economy_GDP_Per_Capita, data_2017.Generosity, color = 'pink')
plt.xlabel('Economy_GDP_Per_Capita')
plt.ylabel('generosity')
plt.title('GDP per capita vs generosity')

plt.figure(figsize=(10,5))
plt.plot(data_2017.Economy_GDP_Per_Capita, data_2017.Generosity, color = 'pink')
plt.xlabel('Economy_GDP_Per_Capita')
plt.ylabel('generosity')
plt.title('GDP per capita vs generosity')

```
